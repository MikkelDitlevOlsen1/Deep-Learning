```python
def load_data(test=False,Print=False,path=f'{os.path.abspath(os.curdir)}/data/carseg_data/clean_data',nr_img=1498):
    train_data_input=[]
    train_data_target=[]
    
    for n in range(nr_img):
        n1=n
        if test:
            n1=f'{n}_a'
        try:
            test1=np.load(f'{path}/{n1}.npy')
            inputs=test1[:3].transpose()
            #inputs=tf.convert_to_tensor(inputs)
            #inputs=inputs[tf.newaxis,...]
            target=test1[3].transpose()
            target=target[...,np.newaxis]
            #target=tf.convert_to_tensor(target)
            #target=target[...,tf.newaxis]
            #target=target[tf.newaxis,...]

            
            train_data_input.append(inputs)
            train_data_target.append(target)
        except:
            if Print:
                print(f'fil nr {n} mangeler')
    return (train_data_input,train_data_target)
```

```python
# Functions to read image and mask paths to output numpy arrays
def read_image(path):
    path = path.decode()
    x = cv2.imread(path, cv2.IMREAD_COLOR)
    x = cv2.resize(x, (IMAGE_SIZE, IMAGE_SIZE))
    x = x/255.0
    return x

def read_mask(path):
    path = path.decode()
    x = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    x = cv2.resize(x, (IMAGE_SIZE, IMAGE_SIZE))
    x = np.expand_dims(x, axis=-1)
    return x
```

```python
# Function to generate tensorflow dataset 
def tf_parse(x, y):
    def _parse(x, y):
        x = x/255.0
    #    x = read_image(x)
    #    y = read_mask(y)
        return x, y

    x, y = tf.numpy_function(_parse, [x, y], [tf.float64, tf.uint8])
    x.set_shape([IMAGE_SIZE, IMAGE_SIZE, 3])
    y.set_shape([IMAGE_SIZE, IMAGE_SIZE, 1])
    return x, y

num_threads = 4

def tf_dataset_train(x, y, batch=batch):
    dataset = tf.data.Dataset.from_tensor_slices((x, y))
    dataset = dataset.shuffle(buffer_size=5000)
    dataset = dataset.repeat()
    dataset = dataset.map(tf_parse, num_parallel_calls=num_threads)
    dataset = dataset.batch(batch)
    #dataset = dataset.map(lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=tf.data.AUTOTUNE)
    dataset = dataset.prefetch(1)
    return dataset


def tf_dataset_valid(x, y, batch=batch):
    dataset = tf.data.Dataset.from_tensor_slices((x, y))
    dataset = dataset.shuffle(buffer_size=500)
    dataset = dataset.repeat()
    dataset = dataset.map(tf_parse, num_parallel_calls=num_threads)
    dataset = dataset.batch(batch)
    dataset = dataset.prefetch(1)
    return dataset
```

```python
train_dataset = tf_dataset_train(train_x, train_y)
valid_dataset = tf_dataset_valid(valid_x, valid_y)
test_dataset = tf_dataset_valid(test_x, test_y)
```

```python
# Function for conv2d_block (to be used for building decoder of unet)
def conv2d_block(input_tensor, n_filters, kernel_size = 3, batchnorm = True):
    x = Conv2D(filters = n_filters, kernel_size = (kernel_size, kernel_size), kernel_initializer = 'he_normal', padding = 'same')(input_tensor)
    if batchnorm:
        x = BatchNormalization()(x)
    x = Activation('relu')(x)
    return x
```

```python
# Function for unet model creation
def unet(n_filters = 16, batchnorm = True, dropout = 0.1, num_classes=num_classes):
    '''
    # Contracting Path (encoder)
    inputs = Input(shape=(IMAGE_SIZE, IMAGE_SIZE, 3), name="input_image")
    encoder = MobileNetV2(input_tensor=inputs, include_top=False)

    BASE_WEIGHT_PATH = ('https://github.com/fchollet/deep-learning-models/releases/tag/v0.6')
    model_name = 'mobilenet_%s_%d_tf_no_top.h5' % ('1_0', 256) #changed 224 to 256
    weight_path = BASE_WEIGHT_PATH + model_name
    weights_path = keras.utils.get_file(model_name, weight_path)
    encoder.load_weights(weights_path, by_name=True, skip_mismatch=True)

    skip_connection_names = ["input_image", "block_1_expand_relu", "block_3_expand_relu", "block_6_expand_relu"]
    encoder_output = encoder.get_layer("block_13_expand_relu").output

    x = encoder_output
    x_skip_1 = encoder.get_layer(skip_connection_names[-1]).output # 224x224 256
    x_skip_2 = encoder.get_layer(skip_connection_names[-2]).output # 112x112 128
    x_skip_3 = encoder.get_layer(skip_connection_names[-3]).output # 56x56 64
    x_skip_4 = encoder.get_layer(skip_connection_names[-4]).output # 28x28 32
    '''
    
    #encoder
    
    inputs = Input(shape=(IMAGE_SIZE, IMAGE_SIZE, 3), name="input_image")
    
    x4  = Input(shape=(IMAGE_SIZE, IMAGE_SIZE, 3), name="input_image") 
    x4 = Conv2D(3, (3,3), strides=(2,2), padding='same')(x4)
    x4 = BatchNormalization()(x4)
    x_skip_4 = Activation('relu')(x4)
    
    x3 = Conv2D(96, (3,3), strides=(2,2), padding='same')(x_skip_4)
    x3 = BatchNormalization()(x3)
    x_skip_3 = Activation('relu')(x3)
    
    x2 = Conv2D(144, (3,3), strides=(2,2), padding='same')(x_skip_3)
    x2 = BatchNormalization()(x2)
    x_skip_2 = Activation('relu')(x2)
    
    x1 = Conv2D(192, (3,3), strides=(2,2), padding='same')(x_skip_2)
    x1 = BatchNormalization()(x1)
    x_skip_1 = Activation('relu')(x1)
    
    x = Conv2D(576, (3,3), strides=(2,2), padding='same')(x_skip_1)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    

    # Expansive Path (decoder)
    u6 = Conv2DTranspose(n_filters * 13, (3, 3), strides = (2, 2), padding = 'same')(x) #32*32*208
    u6 = concatenate([u6, x_skip_1]) #64*64*192, 64*64*144
    c6 = conv2d_block(u6, n_filters * 13, kernel_size = 3, batchnorm = batchnorm) #64*64*336, 64*64*192
    p6 = Dropout(dropout)(c6) #64*64*192 activation

    u7 = Conv2DTranspose(n_filters * 12, (3, 3), strides = (2, 2), padding = 'same')(c6) #64*64*192
    u7 = concatenate([u7, x_skip_2])#128*128*176, 128*128*96
    c7 = conv2d_block(u7, n_filters * 12, kernel_size = 3, batchnorm = batchnorm) #128*128*176, 128*128*176
    p7 = Dropout(dropout)(c7)

    u8 = Conv2DTranspose(n_filters * 11, (3, 3), strides = (2, 2), padding = 'same')(p7) #64*64*192
    u8 = concatenate([u8, x_skip_3]) #128*128*176, 128*128*96
    c8 = conv2d_block(u8, n_filters * 11, kernel_size = 3, batchnorm = batchnorm) #128*128*176
    p7 = Dropout(dropout)(c8)

    u9 = Conv2DTranspose(n_filters * 10, (3, 3), strides = (2, 2), padding = 'same')(c8) #128*128*176
    u9 = concatenate([u9, x_skip_4]) #256*256*160, 256*256*3
    c9 = conv2d_block(u9, n_filters * 10, kernel_size = 3, batchnorm = batchnorm)  #256*256*160
    
    outputs = Conv2D(num_classes, (1, 1), activation='softmax')(c9) #256*256*8

    model = Model(inputs=inputs, outputs=outputs)
    
    return model
```

```python
# Functions for dice coefficient and dice loss 

import tensorflow.keras.backend as K

def dice_coef(y_true, y_pred, smooth=1e-7):
    y_true_f = K.flatten(K.one_hot(K.cast(y_true, 'int32'), num_classes=151)[Ellipsis,1:])
    y_pred_f = K.flatten(y_pred[...,1:])
    intersect = K.sum(y_true_f * y_pred_f, axis=-1)
    denom = K.sum(y_true_f + y_pred_f, axis=-1)
    return K.mean((2. * intersect / (denom + smooth)))

def dice_loss(y_true, y_pred):
    return 1 - dice_coef(y_true, y_pred)
```

```python
# Set up loss function, optimiser and complile the model 
opt = tf.keras.optimizers.Adam(LR)
metrics = [dice_coef, 'accuracy']
model.compile(loss=dice_loss, optimizer=opt, metrics=metrics)
#model.compile(optimizer=opt, loss = tf.keras.losses.SparseCategoricalCrossentropy(),metrics=['accuracy'])
```

```python
# Set up early stopping and reduce learning rate on plateau
early_stopping = EarlyStopping(
    monitor='val_accuracy', 
    patience=3, 
    mode='max'
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_accuracy', 
    factor=0.2,   
    patience=2, 
    min_lr=0.00001,
)
```
